```python
df=df.set_index('Lat')
```

```python
df.head()
```

```python
df1 = df[['Cloudiness','Humidity','Max Temp','Wind Speed']]
df1.head()
```

```python
#graphs
for i in df1:
    plt.scatter(df1.index, df1[i])
    plt.title(f'Lat Vs {i}')
    plt.grid()
    plt.xlabel('Lat')
    plt.ylabel(i)
    plt.show()
    plt.savefig(f'Lat Vs.{i}.png')
```
